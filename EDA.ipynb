```python
import pandas as pd


year2023StatForWorldsTeam = pd.read_excel(r"year2023StatForWorldsTeam.xlsx") #source: https://oracleselixir.com/stats/teams/byTournament

year2023StatForWorldsTeam
```

```python
round3WinLose = pd.read_excel(r"Round3WinLose.xlsx") # source: https://lol.fandom.com/wiki/League_of_Legends_Esports_Wiki

round3WinLose
```

```python
Team1vsTeam2 = pd.DataFrame(columns = list(year2023StatForWorldsTeam.columns.values)) # Create new dataset for the secondary analysis
Team1vsTeam2 = Team1vsTeam2.drop(columns=['team'])
for i in range(len(round3WinLose)):
    team1 = round3WinLose.loc[i][0] # Find the name for Team1
    team2 = round3WinLose.loc[i][1] # Find the name for Team2
    team1data = year2023StatForWorldsTeam.loc[year2023StatForWorldsTeam["team"] == team1].set_index('team') # Get the team1 data by the name we extract
    team2data = year2023StatForWorldsTeam.loc[year2023StatForWorldsTeam["team"] == team2].set_index('team') # Get the team2 data by the name we extract
    compdata = team1data.sub(team2data.squeeze())      # Let team1 data minus team2 data, find the gap between 2 teams
    Team1vsTeam2 = pd.concat([Team1vsTeam2, compdata]) # Add to our Analysis dataset
    
Team1vsTeam2 = Team1vsTeam2.reset_index()     
Team1vsTeam2['IsWin'] = round3WinLose['IsTeam1Won']
Team1vsTeam2
```

```python
 # 删除三列并返回新的数据框
df = Team1vsTeam2.drop(columns=['GP', 'W', 'L'])
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (15,9))
sns.heatmap(df.iloc[:, 1:-1].corr(), annot = True, cmap = 'RdBu_r')
plt.title("Pearson Correlation")
plt.show()
```

```python
df1 = df.drop(columns=['df_index'])

# 创建箱线图
plt.figure(figsize=(15, 9))
sns.boxplot(data=df1)
plt.title("Boxplots for DataFrame Columns")
plt.xticks(rotation=45)
plt.show()
```
